## 1.1 - Why is this relevant? <a class="anchor" id="bullet1x1"></a>
##### [Back to TOC](#TOC)

A basic concept in linguistics and biblical interpretation is the fundamental idea that syntax (the relationships between words) determines meaning more than morphology alone. Hence, how words are used in context is most determinative than their individual meaning. The view is that there are multiple "layers" of language.[^1] There are situations where the form of a word does not match what we expect for its function . Sometimes there are exceptions for poetic reasons, emphasis, or idiom. The overall flow and logic of the passage determine function, not rigid rules of form.

The Formal Morphological tag classifies the smallest meaningful textual units (called morphemes) and the tags provide information about the grammatical features and properties of the individual words. In Greek, these tags can indicate the word's tense, voice, mood, number, gender, case, and other similar attributes.

Functional Tags, on the other hand, focus on the syntactic or functional role of words within a sentence and are  particularly important for parsing and analyzing sentence structure, as they provide insights into the roles of words in phrases, clauses, and larger syntactic units.

## 3.1 - Table with differences between formal and functional tag <a class="anchor" id="bullet3_1"></a>
##### [Back to TOC](#TOC)

The following code will determine the first few Greek words where form and function are considered to be different. The first example will obtain the results in the form of a table. The output is limited to the first 10 occurences.

```python
max=10
count=0
for node in F.otype.s('word'):
    functionaltag=F.functionaltag.v(node)
    formaltag=F.formaltag.v(node)
    if formaltag != functionaltag:
        count+=1
        book=F.book.v(node)
        chapter=F.chapter.v(node)
        verse=F.verse.v(node)
        word=F.word.v(node)
        gloss=F.gloss_EN.v(node)
        print (count,' - ',book,chapter,':',verse,'=\t',word,'\tformaltag:',formaltag,'\tfunctionaltag:',functionaltag, '\tgloss:', gloss)
    if count==max: break
    
```

# 3.2 - Frequency of changes between form and function <a class="anchor" id="bullet3x2"></a>
##### [Back to TOC](#TOC)


```python
# Library to format table
from tabulate import tabulate

ResultDict = {}
for node in F.otype.s('word'):
    FunctionalTag=F.functionaltag.v(node)
    FormalTag=F.formaltag.v(node)
    if FormalTag != FunctionalTag:
        Move=FormalTag+" -> "+FunctionalTag
        # Check if this Move already exists in ResultDict
        if Move in ResultDict:
            # If it exists, add the count to the existing value
            ResultDict[Move]+=1
        else:
            # If it doesn't exist, initialize the count as the value
            ResultDict[Move]=1
            
# Convert the dictionary into a list of key-value pairs and sort it according to frequency
UnsortedTableData = [[key, value] for key, value in ResultDict.items()]
TableData= sorted(UnsortedTableData, key=lambda row: row[1], reverse=True)

# In this example the table will be truncated 
max_rows = 10  # Set your desired number of rows here
TruncatedTable = TableData[:max_rows]

# Produce the table
headers = ["Form -> function","frequency"]
print(tabulate(TruncatedTable, headers=headers, tablefmt='fancy_grid'))

# Add a warning using markdown (API call A.dm) allowing it to be printed in bold type
N1904.dm("**Warning: table truncated!**")
```

## 3.4 - Present Tense as History or Future<a class="anchor" id="bullet3x4"></a>
##### [Back to TOC](#TOC)

Merkle point to two examples where the Greek word ἔρχεται (in its form a preasens) is being used to subsequently indicate a historical  and a future meaning. In this section, these verses will be examined in respect to how the Text-Fabric presents the formal and functional tag. 

According to Merkle:
> *Historical* (a past event that adds vividness or give literary prominence to some aspect of the story): "And the lepper <u>came</u> [ἔρχεται] to Him" (Mark 1:40).

> *Futuristic* (an action that will occur in the future: "Behold, he <u>is comming</u> [ἔρχεται] with the clouds" (Rev. 1:7).[1] 

[1] Benjamin L. Merkle, *Exegetical gems from Biblical Greek*  (Grand Rappids, MI: Baker Academic, 2019), 58.


## 3.4.1  - Table translation of ἔρχεται<a class="anchor" id="bullet3x4x1"></a>
##### [Back to TOC](#TOC)

First examine all uses of ἔρχεται in connection to the formal and functional tag as found in the Text-Fabric dataset together with the associated gloss.

```python
# Load library to format table
from tabulate import tabulate

# Define and execute the query
ErchetaiQuery = '''
book
 chapter
   verse
      word word=ἔρχεται
'''

# The following will create a list containing ordered tuples consisting of node numbers of the items as they appear in the query
ErchetaiResults = N1904.search(ErchetaiQuery)

# Enrich the results with additional feature data
TableData = []
for results in ErchetaiResults:
    # result is a tuple consisting of book, chapter, verse and word
    WordNode=results[3]
    Verse=N1904.sectionStrFromNode(WordNode)
    Gloss=F.gloss_EN.v(WordNode)
    FunctionalTag=F.functionaltag.v(WordNode)
    FormalTag=F.formaltag.v(WordNode)
    TableEntry=[Verse, FunctionalTag, FormalTag, Gloss]
    TableData.append(TableEntry)

# In this example the table will be truncated 
max_rows = 10  # Set your desired number of rows here
TruncatedTable = TableData[:max_rows]

# Produce the table
headers = ["Verse","Formal. tag","Funct. Tag","Gloss"]
print(tabulate(TruncatedTable, headers=headers, tablefmt='fancy_grid'))

# Add a warning using markdown (API call A.dm) allowing it to be printed in bold type
N1904.dm("**Warning: table truncated!**")
```

## 3.5 - Comments <a class="anchor" id="bullet3x5"></a>
##### [Back to TOC](#TOC)

The table in [section 3.2 (Frequency of changes between form and function)](#bullet3x2) shows the differences between formal and functional tag only being pressent for verbs. This matches the remark found on [biblicalhumanities/Nestle1904/morph/](https://github.com/biblicalhumanities/Nestle1904/tree/master/morph):
> Two morphological tags are provided: One which is more "functional", and one which is more "form-oriented". The differences between the two only exist in the description of the verbal system. The form-oriented system is very careful not to make morphological distinctions which are not warranted by the form. For example, the present indicative does not exhibit any formal characteristics which allows the distinction between the middle and the passive voice, hence present indicative verbs are always described as being middle/passive in the "form-oriented" tag. The "functional" tag, on the other hand, sometimes makes a distinction between the middle and the passive, even in the tenses where this is not formally distinguished by the forms. This is so as to be sensitive to the semantics of the verb.
