# 3 - Performing the queries <a class="anchor" id="third-bullet"></a>

## 3.1 - Try <a class="anchor" id="bullet3_1"></a>
##### [Back to TOC](#TOC)

Query

```python
SearchMismatches = '''
phrase
      a:word sp=prep
      <: b:word sp=adj
      <: c:word sp=noun 
a .gn#gn. b 
b .gn#gn. c
a .case#case. b
b .case#case. c
'''
MismatchesList = N1904.search(SearchMismatches)
```

The resulting data (stored in WorshipList) can be further processed. For example to print the first 5 occurences in a table:


```python
N1904.table(MismatchesList, condensed=True)
```
